```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

data = pd.read_csv('./sensor.csv').drop('Unnamed: 0',axis=1)
data
```

```python
data.isnull().sum()
```

The value of the sensor_15 column is completely empty, and it is directly eliminated. Null values of other columns are filled backwards.

```python
data.drop('sensor_15',axis=1,inplace=True)
data.fillna(method='ffill',inplace=True)
data.info()
```

```python
X = data.drop(['timestamp','machine_status'],axis=1)
y = data['machine_status']
```

```python
# Use PCA to reduce to 32 dimensions
from sklearn.decomposition import PCA

n_components = 32
X_pca = PCA(n_components = n_components).fit_transform(X)

print('X_pca.shape',X_pca.shape)
```

```python
X_pca
```

```python
y = y.map({'NORMAL':0,'RECOVERING':1,'BROKEN':2})
y.value_counts()
```

The categories are extremely unbalanced and the training effect may be poor.

```python
# Divide training and test sets
from sklearn.model_selection import train_test_split

x_pca_train,x_pca_test,y_train,y_test = train_test_split(X_pca,y,test_size=0.2,random_state=11,stratify=y)
```

```python
# One-Hot
import keras

y_train = keras.utils.to_categorical(y_train, 3)
y_test = keras.utils.to_categorical(y_test, 3)
```

```python
#  Build a full neural network
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop

model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(32,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model.fit(x_pca_train, y_train,
                    batch_size=32,
                    epochs=10,
                    verbose=1,
                    validation_data=(x_pca_test, y_test))
```

```python
score = model.evaluate(x_pca_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
